# Open dataframe

```python
dataframe_path = './data/COVID-19 Global Statistics Dataset.csv'

df = pd.read_csv(dataframe_path)
```

```python
def describe_column(df, column):
    print(column.upper())
    print("#"*30)
    print(df[column].describe())
    print("\n")
    
for column in df.columns:
    describe_column(df, column)
```
